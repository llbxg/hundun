```python
import sys
from pathlib import Path
sys.path.append(str(Path('__file__').resolve().parent.parent))
```

```python
import numpy as np

from hundun.utils._draw import Drawing
from hundun.exploration._afn import calc_afn
from hundun.exploration._fnn import calc_fnn
```

```python
u_seq = np.load('data/lorenz_x.npy')
```

```python
percentage_list = calc_fnn(u_seq, D=100)
why, what = calc_afn(u_seq, D=1)
```

```python
d = Drawing()
for label, data in [('FNNP', percentage_list), ('E1', why), ('E2', what)]:
    d[0,0].plot(range(1, len(data)+1), data, 
                marker='o', markersize=5, label=label)
d[0,0].set_axis_label('dimension', 'FNNP, ~E1, ~E2')
d[0,0].legend(loc='center right')
d[0,0].set_xlim(0.9, 10.1)
d.show()
```
