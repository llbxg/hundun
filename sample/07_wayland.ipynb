```python
import sys
from pathlib import Path
sys.path.append(str(Path('__file__').resolve().parent.parent))
```

```python
import numpy as np

from hundun.exploration._wayland import wayland
from hundun.utils._draw import Drawing
```

```python
u_seq1 = np.load('data/lorenz_x.npy')[:1024*5]
u_seq2 = np.load('data/henon_xy.npy')[:1024*5, 0]
```

```python
d = Drawing()

for label, u_seq in [('lorenz $x$', u_seq1), ('henon $x$', u_seq2)]:
    median_e_trans_ave = wayland(u_seq)
    d[0,0].plot(
        range(1, len(median_e_trans_ave)+1), median_e_trans_ave,
        label=label)

    d[0,0].scatter(median_e_trans_ave.argmin()+1, median_e_trans_ave.min())

d[0,0].set_yscale('log')
d[0,0].legend()
d[0,0].set_axis_label('dimension', 'median(E_{trans})')
d.show()
```
